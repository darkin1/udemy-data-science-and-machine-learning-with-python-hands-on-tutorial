# Iris SVC

```python
%matplotlib inline

import pandas as pd
from sklearn import svm
from pylab import *
from sklearn.decomposition import PCA


data = pd.read_csv('/home/dariusz/Dokumenty/DataScience-Python3/dc_dataset/iris.data.csv')
data.head()

dt = pd.read_csv('/home/dariusz/Dokumenty/DataScience-Python3/dc_dataset/iris.data.csv', usecols=['sepalLength','sepalWidth','petalLength','petalWidth'])
dt.values
```

```python
d = {'Iris-setosa' : 1, 'Iris-versicolor' : 2, 'Iris-virginica' : 3}
iris_class = data['type'].map(d) 
C = 1.0
classifier = svm.SVC(kernel='linear', C=C, probability=True).fit(dt.values, iris_class.values)
```

# Another way to do it using PCA

```python
pca = PCA(n_components=2, whiten=True).fit(dt)
X_pca = pca.transform(dt)
#print(pca.components_)
print(X_pca)
# print(X_pca[iris_class == 1, 0])
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```
